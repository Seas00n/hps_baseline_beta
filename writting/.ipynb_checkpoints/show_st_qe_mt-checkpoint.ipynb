```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sys
sys.path.append("../")
# from hps_baseline.greggdataset_loader import load_data, v_key, s_key
from hps_baseline.opensource_loader import load_data, v_key, s_key
from scipy.ndimage import gaussian_filter1d
from scipy.optimize import curve_fit
from scipy.signal import argrelextrema

data1 = load_data(vk="0.8", sk="0",m=70)
q_k1 = gaussian_filter1d(data1[0],2)
t_k1 = gaussian_filter1d(data1[1],2)/70
q_a1 = gaussian_filter1d(data1[2],2)
t_a1 = gaussian_filter1d(data1[3],2)/70

data2 = load_data(vk="1.0", sk="0",m=70)
q_k2 = gaussian_filter1d(data2[0],2)
t_k2 = gaussian_filter1d(data2[1],2)/70
q_a2 = gaussian_filter1d(data2[2],2)
t_a2 = gaussian_filter1d(data2[3],2)/70

data3 = load_data(vk="1.2", sk="0",m=70)
q_k3 = gaussian_filter1d(data3[0],2)
t_k3 = gaussian_filter1d(data3[1],2)/70
q_a3 = gaussian_filter1d(data3[2],2)
t_a3 = gaussian_filter1d(data3[3],2)/70


idx = np.arange(q_k1.shape[0])
# color_list = ["#A51C36","#84BA42","#4485C7","#682487"]
color_list = ['r', 'g', 'b', 'm']
```

```python
def cal_st_qe_MT(q_k, t_k, q_a, t_a, ax1, ax2, lstyle, alpha, label):
    def staticimpFun(x, k, qe):
        return k*(x-qe)
    def cal_qe_knee_phase1():
        p_1 = np.argmax(t_k[0:25])
        p_0 = 0
        imp_k,pcov = curve_fit(staticimpFun, q_k[p_0:p_1],t_k[p_0:p_1], 
                               bounds=((0,0),(100,40)))
        x_ = np.arange(q_k[0], q_k[p_1])
        y_ = staticimpFun(x_, imp_k[0], imp_k[1])
        ax1.plot(x_, y_, linestyle=lstyle, color=color_list[0], alpha=alpha,
                 label="Midstance: "+label)
    cal_qe_knee_phase1()

    def cal_qe_knee_phase2():
        p_1 = argrelextrema(t_k[30:60], np.less)[0][0]+30
        p_2 = p_1+10
        imp_k,pcov = curve_fit(staticimpFun, q_k[p_1:p_2],t_k[p_1:p_2], 
                               bounds=((0,0),(5,30)))
        x_ = np.arange(q_k[p_1], q_k[p_2]+10)
        y_ = staticimpFun(x_, imp_k[0], imp_k[1])
        ax1.plot(x_, y_, linestyle=lstyle, color=color_list[1], alpha=alpha,
                 label="Late Stance: "+label)
    cal_qe_knee_phase2()

    def cal_qe_ankle_phase1():
        p_1 = np.argmin(q_a[0:15])
        p_2 = p_1 + 15
        imp_a,_ = curve_fit(staticimpFun, q_a[p_1:p_2],
                        t_a[p_1:p_2], bounds=((0,-15),(7,15)))
        x_ = np.arange(q_a[p_1], q_a[p_2]+10)
        y_ = staticimpFun(x_, imp_a[0], imp_a[1])
        ax2.plot(x_, y_, linestyle=lstyle, color=color_list[0], alpha=alpha,
                 label="Midstance: "+label)
    cal_qe_ankle_phase1()
    
    def cal_qe_ankle_phase2():
        p_1 = np.argmax(t_a)
        p_2 = p_1 + 20
        imp_a,_ = curve_fit(staticimpFun, q_a[p_1:p_2], t_a[p_1:p_2], bounds=((0,-5),(15,15)))
        x_ = np.arange(q_a[p_2], q_a[p_1])
        y_ = staticimpFun(x_, imp_a[0], imp_a[1])
        ax2.plot(x_, y_, linestyle=lstyle, color=color_list[1], alpha=alpha,
                 label="Late Stance: "+label)
    cal_qe_ankle_phase2()       
```

```python
fig = plt.figure(figsize=(10,5))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

plt.style.use('seaborn-v0_8-paper')
ax1.plot(q_k1, t_k1, color='black', alpha=0.2,linewidth=5)
ax1.plot(q_k2, t_k2, color='black', alpha=0.2,linewidth=5)
ax1.plot(q_k3, t_k3, color='black', alpha=0.2,linewidth=5)
ax1.set_ylabel('Knee Torque (N-m/Kg)', fontdict={'family' : 'Times New Roman', 'size': 10})
ax1.set_xlabel('Knee Joint (Deg)', fontdict={'family' : 'Times New Roman', 'size': 10})


ax2.plot(q_a1, t_a1, color='black', alpha=0.2,linewidth=5)
ax2.plot(q_a2, t_a2, color='black', alpha=0.2,linewidth=5)
ax2.plot(q_a3, t_a3, color='black', alpha=0.2,linewidth=5)
ax2.set_ylabel('Ankle Torque (N-m/Kg)', fontdict={'family' : 'Times New Roman', 'size': 10})
ax2.set_xlabel('Ankle Angle (Deg)', fontdict={'family' : 'Times New Roman', 'size': 10})

cal_st_qe_MT(q_k1, t_k1, q_a1, t_a1, ax1, ax2, lstyle=(0,(5,5)),alpha=1, label='v=0.8m/s')
cal_st_qe_MT(q_k2, t_k2, q_a2, t_a2, ax1, ax2, lstyle=(0,(5,1)),alpha=1, label='v=1.0m/s')
cal_st_qe_MT(q_k3, t_k3, q_a3, t_a3, ax1, ax2, lstyle=(0,(3,1,1,1)),alpha=1, label='v=1.2m/s')
ax1.legend()
ax2.legend()
plt.savefig("../image/st_qe_mt.png")
```
